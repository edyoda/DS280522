```python
# Data Analysis with .groupby and .agg : user-defined-function and lambda fn

# create a statistical table which answers the following questions for every store:

# group by store_id
  # Q1) number of products starting with letter 'A'
  # Q2) product list for every store (example : 2 --> B01 , B02 , B03)
  # Q3) Difference of max price - min price 
  # Q4) First percentile for product price

df = pd.DataFrame(dict(StoreID=[1,1,1,1,2,2,2,3,3,3],
                       ProductID=['A01', 'A02', 'B01','C01',  
                                  'B01', 'B02', 'B03', 'C01', 
                                  'C02', 'C03'],
                       price=[10,20,50,100,50,60,70,100,150,200]))

df

# Q1) number of products starting with letter 'A'

def cnt_A(x): # x --> ProductID string
  counter = 0
  for item in x:
    if item[0] == 'A' : # checking the first index of string
      counter += 1
  return counter


# Q2) product list for every store

# lambda x : ', '.join(x)

# Q3) Difference of max price - min price
#  lambda x : x.max() - x.min()

# Q4) First percentile for product price
    # lambda x : np.percentile(x , 25)




df


```

```python
df.groupby(['StoreID'])\
  .agg(
      count_products_A = ('ProductID' ,cnt_A ),
       product_list = ('ProductID' , lambda x : ', '.join(x)),
       price_diff = ('price' , lambda x : x.max() - x.min()),
       price_q1 = ( 'price', lambda x : np.percentile(x , 25))
  )
```

```python

# transform : used to append statiscal measure to original df

df['count_products_A'] = df.groupby(['StoreID'])['ProductID'].transform(cnt_A)


```

```python
df['first_quantile'] = df.groupby(['StoreID'])['price'].transform(lambda x : np.percentile(x , 25))
```

```python
# Technique 1

grouped_df = df.groupby('Branch')

# df --> groupby --> .agg() --> .transform()

# col 1 --> agg fn mean --> transform

df[['tax_branch_mean' , 'unit_price_mean']]  =   grouped_df[['Tax 5%' ,'Unit price' ]].transform('mean')

# 1. Group by operation on branch col
# 2. grouped rows are stored in grouped_df
# 3. Apply agg fn mean on grouped_df on two columns
# 4. Store the stas in a new col
```

```python
df[['product_line_count' , 'gender_count']]  =   grouped_df[['Product line' ,'Gender' ]].transform('count')
```
